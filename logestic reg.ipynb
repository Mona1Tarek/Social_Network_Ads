```python
data = data.get_dummies()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#We are setting up a tool to convert text labels into numerical representations that machine learning models can understand.

data['Gender'] = le.fit_transform(data['Gender'])
data.head()
```

```python
x = data.drop(['Purchased'], axis=1)
y = data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# x - (mean / std)   --> 3lshan a22lel el range f y2el el computing power
#el standerscalar bt3ml kda 3la kolo 3loa tool
# bn3ml lel x bs msh lel label

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import  LogisticRegression
log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
y_pred = log.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)

#In your case, since you are working with the "Purchased" column, which is a binary variable (0 or 1), you are dealing with a classification problem. Therefore, accuracy_score would be the appropriate metric to use.
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
